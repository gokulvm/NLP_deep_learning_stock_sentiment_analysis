## Data Cleaning

```python
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
test_data = test.iloc[:,2:27]
test_data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
count = [str(i) for i in range(0,25)]
data.columns = count
for i in count:
    data[i] = data[i].str.lower()
    
data.head(2)
```

```python
count = [str(i) for i in range(0,25)]
test_data.columns = count
for i in count:
    test_data[i] = test_data[i].str.lower()
    
test_data.head(2)
```

```python
headline = []
for i in range(0,len(data.index)):
    sen = " ".join(str(j) for j in data.iloc[i,0:25])
    headline.append(sen)
    
headline[0]
```

```python
test_headline = []
for i in range(0,len(test_data.index)):
    sen = " ".join(str(j) for j in test_data.iloc[i,0:25])
    test_headline.append(sen)
    
test_headline[100]
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

ps = PorterStemmer()
wl=WordNetLemmatizer()


stem = []
for i in range(0,len(headline)):
    new = headline[i]
    new = new.split()
    new = [ps.stem(word) for word in new if not word in set(stopwords.words("english"))]
    new = " ".join(new)
    stem.append(new)
```

```python
test_stem = []
for i in range(0,len(test_headline)):
    new = test_headline[i]
    new = new.split()
    new = [ps.stem(word) for word in new if not word in set(stopwords.words("english"))]
    new = " ".join(new)
    test_stem.append(new)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
countvector=CountVectorizer(ngram_range=(2,2),max_features=500000)
traindataset=countvector.fit_transform(stem)
traindataset
```

```python
testdataset=countvector.transform(test_stem)
testdataset
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn import metrics
import matplotlib as plt
import seaborn as sns
```

```python
def prediction(x_test,y_test,model,algo):        
        print(f"Predictions for {algo} Algorithem")
        print("Test Results: ")
        print("================")
        pre_t = model.predict(x_test)
        print("        Accuracy Score    : {:.2f} %".format((metrics.accuracy_score(y_test,pre_t))*100))
        print("")        
        print("        F1 Score          : {:.4f} ".format((metrics.f1_score(y_test,pre_t,average="macro"))))
        print("")        
        print("        Precision         : {:.4f} ".format((metrics.precision_score(y_test,pre_t,average="macro"))))
        print("")
        print("        Recall            : {:.4f} ".format((metrics.recall_score(y_test,pre_t,average="macro"))))
        print(     "Classification Report  :  ")
        print(f" {metrics.classification_report(y_test,pre_t)}")        
```

```python
colorm = ["YlOrRd","BuPu","PuBu","YlGn","RdBu"]
import random as ran
def con_mat(x_test,y_test,model,algo):
    print(f"Confusion matrix for {algo} Algorithem")
    pre_t = model.predict(x_test)
    cm_lr = metrics.confusion_matrix(y_test,pre_t)
    sns.heatmap(cm_lr,cmap=ran.choice(colorm),annot=True,cbar=True,annot_kws={"size": 20})
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=16000
onehot_repr = [one_hot(words,voc_size)for words in stem] 
test_onehot_repr=[one_hot(words,voc_size)for words in test_stem] 
```

## Embedding Representation

```python
sent_length= 200
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
sent_length= 200
test_embedded_docs=pad_sequences(test_onehot_repr,padding='pre',maxlen=sent_length)
print(test_embedded_docs)
```

```python
embedded_docs[3000]
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=60
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.5))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),train['Label'].shape
```

```python
y = train['Label']
```

```python
import numpy as np
X_train=np.array(embedded_docs)
y_train=np.array(y)
```

```python
test_y = test['Label']
```

```python
X_test=np.array(test_embedded_docs)
y_test=np.array(test_y)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20,batch_size=32)
```

```python
pre = model.predict(X_test)
```

```python
for i in range(0,len(pre)):
    if pre[i] < 0.5:
        pre[i] = 0
    else:
        pre[i] = 1
```

```python
#pre = pre.reshape(1)
pre = pre.astype("int64")
```

```python
pre = pre.flatten()
```

```python
confusion_matrix(y_test,pre)
```

```python
print("Accuracy Score : {:.2f} %".format((metrics.accuracy_score(y_test,pre))*100))
print("F1 Score : {:.4f} ".format((metrics.f1_score(y_test,pre,average="macro"))))
```

```python
print("Classification Report  :  ")
print(f" {metrics.classification_report(y_test,pre)}")        
```
